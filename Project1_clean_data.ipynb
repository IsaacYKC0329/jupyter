```python
import pandas as pd #读取所有文件

Tokyo = pd.read_csv('Tokyo 2021 medal.csv')

Pop = pd.read_csv('Population copy.csv')

Rio = pd.read_csv('Rio_olympic_Medals.csv')

GDP = pd.read_csv('GDP.csv')

Age = pd.read_csv('Age_1.csv')

Age
```

```python
TOKRIO = pd.merge(Tokyo, Rio, how = 'outer',
                  left_on = 'NOCCode', right_on = 'Country_Code'
                 )
TOKRIO 
#合并tokyo 和 Rio， 以country code 这个相同column 合并， 并取名TOKRIO
```

```python
GDPFinal = GDP.filter(['Country Name','Country Code', '2016','2017','2018','2019' ,'2020'])
GDPFinal
#抽取以上几个column name， 合并成新的csv 叫GDPFinal
```

```python
POPCOD = pd.merge(Pop, TOKRIO, how = 'outer',
                  left_on = 'Country Name', right_on = 'Country_Name'
                 )
POPFinal = POPCOD
POPCOD
#把人口 和TOKRIO 以Country Name进行合并 叫POPFinal
```

```python
GDPFinal_1 = pd.merge(POPFinal, GDPFinal, how = 'outer',
                  left_on = 'Country Name', right_on = 'Country Name'
                 )
GDPFinal_1
#把POPFinal 和 GDPFinal ，以country name 进行合并
```

```python
Age1 = Age.filter(['Country Name', '2020'])
Age1
#提取 老龄率中的 country name 和 2020 
```

```python
Final2 = GDPFinal_1.filter(['Country Name', 'Country Code_x', '2016', '2017','2018','2019', '2020_y','2020_x', 'Total', 'Rio_2016_Total_Medals'])
Final2
#提取'Country Name', 'Country Code_x', '2016', '2017','2018','2019', '2020_y','2020_x', 'Total', 'Rio_2016_Total_Medals'
#合并成新的csv 叫 Final2
```

```python
Final4 = pd.merge(Final2, Age1, how='outer',
                  left_on='Country Name', right_on= 'Country Name'
                 )
Final4
#同上
```

```python
Final4['Total Medals'] = Final4[["Total", "Rio_2016_Total_Medals"]].apply(lambda x:x["Total"]+x["Rio_2016_Total_Medals"], axis=1)
Final4
#提取"Total", "Rio_2016_Total_Medals" 相加， 并成为新的column，叫'Total Medals'


```

```python
Final5 = Final4.rename(columns = {"Country Code_x": "Country Code","2020_x": "Populations", "2020_y": "2020_GDP($)", "2016": "2016_GDP($)", "2017" :"2017_GDP($)", "2018": "2018_GDP($)", "2019": "2019_GDP($)", "Rio_2016_Total_Medals": "Rio Medals", "Total": "Tokyo Medals", "2020": "Old age rate"})
Final5
#改名字
```

```python
B = Final5[['2016_GDP($)', '2017_GDP($)', '2018_GDP($)', '2019_GDP($)', '2020_GDP($)']]
Final5['Average GDP($)'] = B.mean(axis = 1)
Final6 = Final5.dropna()
Final6
#提取'2016_GDP($)', '2017_GDP($)', '2018_GDP($)', '2019_GDP($)', '2020_GDP($)'
#求平均数
```

```python
Final7 = Final6.sort_values(by= ['Total Medals'], ascending = False)
Final7
#sortby
```

```python
Final7.index = range(1, len(Final7)+1)
Final7
#加索引
```

```python
Final7.index.name = "Rank"
Final7
#改索引的名字
```

```python
cols = Final7.columns.tolist()
cols.insert(0, cols.pop(cols.index("Total Medals")))
cols.insert(1, cols.pop(cols.index("Tokyo Medals")))
cols.insert(2, cols.pop(cols.index("Rio Medals")))
cols.insert(10, cols.pop(cols.index("Average GDP($)")))

Final8 = Final7[cols]
Final8
#调整column的位置
```

```python
Final8.to_csv("Project1 Final 注释")
#到处csv file
```
