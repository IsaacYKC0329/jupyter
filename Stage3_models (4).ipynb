# Lesso Model

```python
mu = x_train.mean()
sigma = x_train.std()
```

```python
x_train = (x_train - mu)/sigma
x_test = (x_test - mu)/sigma
```

```python
from sklearn.linear_model import LassoCV
lasso = LassoCV(cv = 5)
lasso.fit(x_train, np.ravel(y_train))
```

```python
pd.DataFrame(lasso.coef_.round(3), index = col_use).T
```

```python
print("Training RMSE: ", RMSE(lasso.predict(x_train),y_train))
print("Test RMSE: ", RMSE(lasso.predict(x_test), y_test))
print("Training R-squared: ", r2_score(lasso.predict(x_train), y_train))
print("Test Squared: ", r2_score(lasso.predict(x_test), y_test))
```
