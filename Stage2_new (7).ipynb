```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('SDG_goal3_clean.csv')
```

# Data cleaning
## Rename the data

```python
df1 = df.rename(columns = {'Maternal mortality ratio': 'MMR', 'Proportion of births attended by skilled health personnel (%)':'PBP', 
               'Infant mortality rate (deaths per 1,000 live births):::BOTHSEX': 'IMR-B', 
                 'Under-five mortality rate, by sex (deaths per 1,000 live births):::BOTHSEX': 'Under-fiveMR-B',
                'Neonatal mortality rate (deaths per 1,000 live births)': 'NMR',
               'Mortality rate attributed to cardiovascular disease, cancer, diabetes or chronic respiratory disease (probability):::BOTHSEX':'MR-diseae-B',
               'Suicide mortality rate, by sex (deaths per 100,000 population):::BOTHSEX': 'Suicide-MR-B',
               'Death rate due to road traffic injuries, by sex (per 100,000 population):::BOTHSEX': 'Death-MR-B',
               'Adolescent birth rate (per 1,000 women aged 15-19 years)':'AdBR',
               'Health worker density, by type of occupation (per 10,000 population)::PHYSICIAN': 'Phy-den',
               'Health worker density, by type of occupation (per 10,000 population)::NURSEMIDWIFE': 'Nurse-den',
               'Health worker density, by type of occupation (per 10,000 population)::PHARMACIST': 'Phar-den',
               'Universal health coverage (UHC) service coverage index': 'UHC-index',
               'Fraction of deaths due to diabetes, among deaths due to cardiovascular disease, cancer, diabetes or chronic respiratory disease': 'Fraction-diabetes',
               'Fraction of deaths due to cancer, among deaths due to cardiovascular disease, cancer, diabetes or chronic respiratory disease': 'Fraction-cancees'})
df1
```

# Pick up the column i want to use

```python
df2 = df1[['Country','Year','Region','MMR','PBP','IMR-B','Under-fiveMR-B','NMR','MR-diseae-B','Suicide-MR-B','Death-MR-B','AdBR','Phy-den','Nurse-den',
          'Phar-den','UHC-index','Fraction-diabetes','Fraction-cancees']]
df2
```
